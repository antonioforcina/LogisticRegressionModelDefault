# CHECKING NORMAL DISTRIBUTION VARIABLES

```r
#checking normality for the 6 variables income,othdebt+carvalue+creddebt+default+yrsed
#starting from income
```

```r
#building function
plotVar <- function(var, breaks=20, label='')
{
    hist(var, freq=FALSE, breaks=breaks, col=rgb(1,1,0,0.4), 
         main=paste('Histogram of', label), xlab=label)
    curve(dnorm(x, mean=mean(var), sd=sd(var)), 
          from=min(var), to=max(var),
          col='red', lw=4, add=TRUE)
}
```

```r
ks <- function(x)
{ set.seed(13); ks.test(x, rnorm(mean=mean(x), sd=sd(x), n=length(x))) }
```

# BUILD MODEL STARTING FROM ALL VARIABLES

# Simple Linear Model

```r
#CREATING A MODEL WITH ALL THE VARIABLES
model3 <- lm(income ~ othdebt+carvalue+creddebt+default+yrsed, IncomeData3)
summary(model3)

```

```r
vif(model3)
```

```r
sigma(model3)*100/mean(IncomeData3$income)
```

```r

ncvTest(model3)
```

# SYNTHETIC MODEL

```r
SyntMod <- lm(log(income) ~ exp(0.3*log(othdebt)) + log(carvalue) + exp(0.3*log(creddebt)) + default + yrsed, data=IncomeData3)
summary(SyntMod)


```

```r

ncvTest(SyntMod)
```

```r
#computing the percentage of residual standard error


sigma(SyntMod)/mean(IncomeData3$income)
```

```r

ncvTest(SyntMod)
```

# log linear model

```r
loglinm <- lm(log(income) ~ othdebt+carvalue+creddebt+default+yrsed, data=IncomeData3)
summary(loglinm)
```

```r
sigma(loglinm)*100/mean(IncomeData3$income)
```

```r
ncvTest(loglinm)
```
